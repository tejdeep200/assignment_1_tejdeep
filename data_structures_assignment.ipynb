#task4

```python
#1
products = ["chips", "biscuits", "coke", "chocolate", "maggi", "rice", "rusk", "juice"]
category_dict = {
    "chips": "snack",
    "biscuits": "snack",
    "coke": "beverage",
    "chocolate": "snack",
    "maggi": "instant_food",
    "rice": "grain",
    "rusk": "snack",
    "juice": "beverage"
}

price_dict = {
    "chips": 20,
    "biscuits": 25,
    "coke": 45,
    "chocolate": 50,
    "maggi": 15,
    "rice": 60,
    "rusk": 30,
    "juice": 35
}
catalog = []

for product in products:
    catalog.append((product, price_dict[product], category_dict[product]))

print("\nCatalog:")
for item in catalog:
    print(item)

```

```python
#2
category_to_products = {}

for product, price, category in catalog:
    if category not in category_to_products:
        category_to_products[category] = []
    category_to_products[category].append(product)

print("\nCategory to Products Mapping:")
print(category_to_products)
```

```python
#3
max_category = max(category_to_products, key=lambda c: len(category_to_products[c]))

print("\nCategory with maximum products:", max_category)
```
